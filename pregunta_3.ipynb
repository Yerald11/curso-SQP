```python
def consultas(input):
  messages = [
    {"role"   : "system",
     "content": context1},
    {"role"   : "user",
     "content": "Eres un especialista biopolimeros, hidrogeles y en análisis de texto científico: ¿Cuál es el objetivo principal del estudio presentado en el abstract?"},
  ]

  prompt = pipeline.tokenizer.apply_chat_template(
          messages,
          tokenize=False,
          add_generation_prompt=True
  )

  terminators = [
      pipeline.tokenizer.eos_token_id,
      pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
  ]

  outputs = pipeline(
      prompt,
      max_new_tokens=256,
      eos_token_id=terminators,
      do_sample=True,
      temperature=0.6,
      top_p=0.9,
  )

  return outputs[0]["generated_text"][len(prompt):]
```

```python
df['Analisis1'] = None

for i in range(3):
  context1 = df["Abstract"][i]
  response = consultas(context1)
  df['Analisis1'][i] = response
```
